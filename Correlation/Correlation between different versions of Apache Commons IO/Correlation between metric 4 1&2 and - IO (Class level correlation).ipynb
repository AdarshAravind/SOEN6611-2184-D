#### This block of the script reads Cyclomatic Code Complexity from the CSV file extracted from JUnit results of seven different versions of Apache Commons IO projects.

#### for loop and if condition at the end of the block is used to read only the numeric values from the mentioned column.

```python
import csv
com = []
with open("C:\\Users\\arvin\\Desktop\\IO Versions\\Metric 1&2 and 4\\commons-io-2.0.csv") as csvDataFile:
    code_complexity1=list(csv.reader(csvDataFile))
with open("C:\\Users\\arvin\\Desktop\\IO Versions\\Metric 1&2 and 4\\commons-io-2.1.csv") as csvDataFile:
    code_complexity1.extend(list(csv.reader(csvDataFile)))
with open("C:\\Users\\arvin\\Desktop\\IO Versions\\Metric 1&2 and 4\\commons-io-2.2.csv") as csvDataFile:
    code_complexity1.extend(list(csv.reader(csvDataFile)))
with open("C:\\Users\\arvin\\Desktop\\IO Versions\\Metric 1&2 and 4\\commons-io-2.3.csv") as csvDataFile:
    code_complexity1.extend(list(csv.reader(csvDataFile)))
with open("C:\\Users\\arvin\\Desktop\\IO Versions\\Metric 1&2 and 4\\commons-io-2.4.csv") as csvDataFile:
    code_complexity1.extend(list(csv.reader(csvDataFile)))
with open("C:\\Users\\arvin\\Desktop\\IO Versions\\Metric 1&2 and 4\\commons-io-2.5.csv") as csvDataFile:
    code_complexity1.extend(list(csv.reader(csvDataFile)))
with open("C:\\Users\\arvin\\Desktop\\IO Versions\\Metric 1&2 and 4\\commons-io-2.6.csv") as csvDataFile:
    code_complexity1.extend(list(csv.reader(csvDataFile)))
    count=1
    for column in code_complexity1:
        if(column[10].isdigit()):
            com.append(int(column[10]))
        else:
            continue
print("Class wise Cyclomatic Code Complexity for multiple versions of Apache Commons IO\n")
print(com)
```

#### This block of code calculates and prints the value of spearman correlation between Statement coverage and Cyclomatic Code complexity. And, spearman correlation between Branch coverage and Cyclomatic Code complexity. The value should be in the range [-1, 1] 
#### The results indicates that there exists a strong positive correlation between both "(Statement coverage, Cyclomatic Code complexity)", and between "(Branch coverage and Cyclomatic Code complexity)".

```python
from scipy.stats import spearmanr
print(spearmanr(state, com)) #Correlation between Statement Coverage & Cyclomatic Code Complexity
print(spearmanr(branch, com)) #Correlation between Branch Coverage & Cyclomatic Code Complexity
```

#### Correlation between Statement coverage and Cyclomatic Code complexity is represented using a scatter plot below.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy, pylab
import plotly.plotly as py
import plotly.tools as tls
plt.scatter(state, com)
plt.xlim(0, 7000)
plt.ylim(0, 300)
plt.yticks(np.arange(0, 300, 50))
plt.xticks(np.arange(0, 7000, 1000))
plt.ylabel("Statement Coverage")
plt.xlabel("Cyclomatic Code Complexity")
plt.title("Cyclomatic Code Complexity vs Statement Coverage")
plt.show()
```
